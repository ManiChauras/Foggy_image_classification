# Day Night classifier

```python
import cv2
import os
import numpy as np
import shutil
```

```python

# Function to classify images as day or night
def classify_day_night(image_path):
    image = cv2.imread(image_path)
    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Calculate the average pixel intensity
    average_intensity = np.mean(gray)
    print(average_intensity)

    # Threshold to classify as day or night
    threshold = 62  # Adjust this threshold based on your images
    is_day = average_intensity > threshold

    return is_day
```

```python
# Path to the folder containing your images
folder_path = r'testing/binary/fog'

# Create "day" and "night" folders
day_folder = os.path.join(folder_path, 'day')
night_folder = os.path.join(folder_path, 'night')

os.makedirs(day_folder, exist_ok=True)
os.makedirs(night_folder, exist_ok=True)

# Iterate through the files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.jpg'):
        image_path = os.path.join(folder_path, filename)
        is_day = classify_day_night(image_path)
        print(is_day)

        # Assign labels and move images to corresponding folders
        label = 'Day' if is_day else 'Night'
        destination_folder = day_folder if is_day else night_folder
        print(destination_folder)

        # Move the image to the appropriate folder
        shutil.move(image_path, destination_folder)
```
